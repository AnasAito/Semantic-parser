```python
def calculate_central_nodes(text_network , max_nodes = -1):

    bc = (nx.betweenness_centrality(text_network,weight='weight'))
    #print(bc)
    
    nx.set_node_attributes(text_network, bc, 'betweenness')
    bc_threshold = sorted(bc.values(), reverse=True)[max_nodes]
    to_keep = [n for n in bc if bc[n] > bc_threshold]
    filtered_network = text_network.subgraph(to_keep)
    return filtered_network

def filter_graph_by_degree(graph ,degree_tresh ):
    unit = len(graph.nodes)
    def filter_node(node):
        return unit*(graph.nodes[node]['degree'])>degree_tresh
    return list(filter(filter_node,list(graph.nodes)))

def f_graph_by_degree(text_network , deg_threshold ):
    unit = len(text_network.nodes)
    to_keep = [n for n in list(text_network.nodes) if unit*(text_network.nodes[n]['degree']) > deg_threshold]
    print(to_keep)
    filtered_network = text_network.subgraph(to_keep)
    
    return filtered_network


def get_connectivity(graph,node_a,node_b):
    
    path = nx.shortest_path(graph,source=node_a,target=node_b , weight='distance')
    pairs_ = list(zip(path, path[1:] + path[:1]))[:-1] ## delete last couple that link terminal nodes
    #print(pairs_)
    weights = [graph[pair_[0]][pair_[1]]['weight'] for pair_ in pairs_]
    #print(weights)
    edge_weight=0
    for w in weights : 
        edge_weight+=1/w
    return 1/edge_weight , path


def get_similarity(graph,topic,topics):

    conn=[]

    for node in graph.nodes : 
        if node in topics:
            pass

        else :
        
            try : 
                w,path =  get_connectivity(graph,topic,node)
                conn.append(w)
                ##conn_index.append(node)
            except :
                conn.append(0)
    return conn

def get_df(graph,topics):
    dict_values={}
   # topics = ['plant','waste','battery','soil']
    for topic in topics : 
        print('-------------')
        print(topic)
        value_list = get_similarity(graph,topic,topics)
        ##normed = (np.array(value_list) - min(value_list))/(max(value_list)-min(value_list))
        dict_values[topic] = value_list 
    filtred = [node for node in graph.nodes if node not in topics]   
    df = pd.DataFrame(data=dict_values).set_index([pd.Index(filtred)])
    return df

def get_clusters(df):
    # Convert DataFrame to matrix
    mat = df.values
    # Using sklearn
    km = KMeans(n_clusters=len(df.columns))
    km.fit(mat)
    # Get cluster assignment labels
    labels = km.labels_
    # Format results as a DataFrame
    results = pd.DataFrame([df.index,labels]).T
    results = results.rename(columns={0: 'index',1:'cat'})
    for i in range(len(list(df.columns))) :
        cluster = results[results['cat']==i]['index'].values
        print('----------------')
        print(len(cluster))
        print(cluster)
      
    return results



def add_communities_v1(graph,df,topics):
    for node in graph.nodes:
        if node in topics:
            group = -1
        else :
            row = df.loc[ node , : ]
            values = list(row)
            max_index = values.index(max(values)) 

            group = max_index
        graph.nodes[node]['group']=group
        
        
def add_communities(graph,df_clusters,topics):
    for node in graph.nodes:
        if node in topics:
            group = -1
        else :
            group = int(results[results['index']==node]['cat'].values[0])
        graph.nodes[node]['group']=group
   
```

```python
def process_dict(dict_):
    min_ = min(dict_.items(), key=lambda x: x[1]) [1]
    max_ = max(dict_.items(), key=lambda x: x[1]) [1]
    ## transform
    return  dict((k, (v-min_)/(max_-min_)) for k, v in dict_.items())


from collections import Counter
dicts = [degree,eign,closness]
proc_dicts = [process_dict(dict_) for dict_ in dicts]
c = Counter()
for d in proc_dicts:
    c.update(d)
sorted_scores = sorted(c.items(), key=lambda kv: kv[1])
print(([elem[0] for elem in sorted_scores[-100:]]))
```

```python
def export_graph(graph_dict):
    data = json_graph.node_link_data(graph_dict['graph'])
    formated_title = graph_dict['title'].replace('/', '')
    path = './graph-ui/src/data/'+formated_title+'.json'
    with open(path, 'w') as outfile:
        json.dump(data, outfile)
    
def write_title(title):
    formated_title = title.replace('/', '')
    with open('./graph-ui/src/data/index.json') as json_file:
        data = json.load(json_file)
       ## print(data['graphs'])
        list_ = data['graphs']
    with open('./graph-ui/src/data/index.json','w') as json_file:
    
        list_ = list_+[formated_title]    
        json.dump({'graphs':list_}, json_file) 
    return 'written'
def write_graph(graph,title,export=True):
    ## write to db             
    if export : 
        export_graph({'title':title, 'graph':graph})
        write_title(title)
    return 'written' 
def plot_graph(graph):
    pl.figure()
    nx.draw_networkx(graph)
    pl.show()   
```
